#Jenis-Jenis Operator Aritmatika

```python
#Penjumlahan

jet_pribadi_deo = 21
jet_pribadi_rizal = 8

total_jet_pribadi = jet_pribadi_deo + jet_pribadi_rizal

print(total_jet_pribadi)
```

```python
#Pengurangan

jet_pribadi_deo = 21
jet_pribadi_rizal = 8

selisih_jet_pribadi = jet_pribadi_deo - jet_pribadi_rizal

print(selisih_jet_pribadi)
```

```python
#Perkalian

jumlah_garasi_jet_deo = 3
jumlah_jet_per_garasi = 7

total_jet_deo = jumlah_garasi_jet_deo * jumlah_jet_per_garasi

print(total_jet_deo)
```

```python
#Pembagian

jet_pribadi_deo = 21
jumlah_anak_yatim = 6

jet_tiap_anak_yatim = jet_pribadi_deo / jumlah_anak_yatim

print(jet_tiap_anak_yatim)
```

```python
#Pembagian

jet_pribadi_deo = 21
jumlah_anak_yatim = 6

jet_tiap_anak_yatim = jet_pribadi_deo // jumlah_anak_yatim

print(jet_tiap_anak_yatim)
```

```python
#Sisa bagi

total_jet_deo = 21
jumlah_anak_yatim = 6

sisa_jet_deo = total_jet_deo % jumlah_anak_yatim

print(sisa_jet_deo)
```
