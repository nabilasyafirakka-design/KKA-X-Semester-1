#Aplikasi Menghitung Luas Persegi Panjang

```python
panjang = int(input("Masukkan panjang persegi panjang (cm) : "))
lebar = int(input("Masukkan lebar persegi panjang (cm) : "))

luas = panjang * lebar

print(f"Luas persegi panjang yaitu {luas} cm^2.")
```

#Aplikasi Menghitung Volume Balok

```python
panjang = int(input("Masukkan panjang balok (cm) :"))
lebar = int(input("Masukkan lebar balok (cm) :"))
tinggi = int(input("Masukkan tinggi balok (cm) :"))

volume = panjang * lebar * tinggi

print(f"volume balok yaitu {volume} cm^3.")
```
